### DS4400: Lab 5: Visualizing Gradient Ascent (Descent)

The gradient is a vector function that transforms the scalar function, f(x,y)
into one that returns vectors point along the direction of the steepest ascent.
It involves partial derivatives of f with respect to x and y. 


```python
# Library imports
# If you get import errors it means you need to install these libraries
# At the command line terminal:  conda install <pkg>


import numpy as np
import mltools as ml

import importlib
importlib.reload(ml)
```

```python
# Initializing an array with a function
# We create a 10x10 array and now the X and Y values are integers from 0...9.

def f(x,y):
    return 5 + x**4 - 2 * y**2 + x * y
    #return np.cos(x) * np.sin(y)

def dfdx(x,y):   
    """ the partial derivative delta f / delta x """
    return 4 * x**3 + y
    #return -1 * np.sin(x) * np.sin(y)

def dfdy(x,y):
    """ the partial derivative delta f / delta y"""
    return -4 * y + x
    #return np.cos(x) * np.cos(y)





```

```python
# Test the surface plot function
# Calculate the surface array

x = np.arange(-2, 2, .1)
y = np.arange(-2, 2, .1)

ml.surface_plot(x, y, f)
```

```python
ml.contour_plot(x, y, f, dfdx, dfdy, gradient_field=True, ascend=(0, -2), alpha = .02)
```

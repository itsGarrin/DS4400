```python
# Library imports

import pandas as pd
```

```python
# Read large dataset
songs = pd.read_csv('spotify_full.csv')
```

```python
# Sample 5 songs

five = songs.sample(n=5)
five


```

```python
# Specify a random seed to sample the same five songs
five = songs.sample(n=5, random_state=123)
five
```

```python
# Alternatively, specify a FRACTION of songs to be sampled
sample_1pct = songs.sample(frac=0.01)
len(sample_1pct)
```

```python
# replace:  False (default): Same song can't be picked twice.   

repeats = five.sample(n=5, replace=True)
repeats
```

```python
# The canonical way to shuffle rows in a dataframe
# Something we'll want to do when creating training and test sets

shuffled = songs.sample(frac=1)
shuffled
```

```python
# How many songs in each genre?
songs.groupby(['track_genre'])['track_genre'].count()
```

```python
# Lets focus on Rock, Pop, or Hip Hop.

rph = songs[songs['track_genre'].isin(['rock', 'pop', 'hip-hop'])]
print("Songs: ",len(rph))
rph = rph.sample(frac=1, ignore_index=True, random_state=314159)  # Automatically resets the index to 0, 1, 2, ....
rph.head(10)
```

```python
# Create hw4 training data data
train = rph.iloc[:2000, 5:]
len(train)
train

train.to_csv('spotify_train.csv', index=False)
```

```python
# Create hw4 TEST data (NOT TO BE USED FOR HYPERPARAMETER TUNING - REPORT PERFORMANCE ONLY)
# Create hw4 test data
test = rph.iloc[2000:, 5:].reset_index().iloc[:, 1:]
len(test)
test
test.to_csv('spotify_test.csv', index=False)
```

### Lab 16: K-Fold Cross Validation

Split the TRAINING data into training and validation, leaving the test-set unchanged.
Use validation data to "test" the model and perform hyperparameter tuning.   This is more legitimate
than using the test data for hyperparameter tuning.  Why?  We are then effectively using test data to train the best model - and are subject again to over-fitting. We lose any objective estimate of model performance.

```python
# Imports

import pandas as pd

```

```python
# Read training data

train = pd.read_csv('spotify_train.csv')
train
```

```python
folds = 10

for f in range(folds):
    train_fold = train[train.index % folds != f]
    valid_fold = train[train.index % folds == f]
    print(f, len(train_fold), len(valid_fold))
    
```
